```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Train_data_feat=pd.read_csv('Hackathon-Master/Keras_Train_Feature.csv')
Test_data=pd.read_csv('Hackathon-Master/Keras_Test.csv')
Train_data_label=pd.read_csv('Hackathon-Master/Keras_Train_Labels.csv')
```

```python
Train_Feat=Train_data_feat.copy()
Train_lab=Train_data_label.copy()
Test_Feat=Test_data.copy()
```

```python
Test_Feat
```

```python
print('The Shape of Train Feature : {} and the Train Label={} and the Test label : {}'.format(Train_Feat.shape, Train_lab.shape, Test_Feat.shape))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from keras import backend as K
import tensorflow as tf
```

```python
X_train, X_test, y_train, y_test = train_test_split(Train_Feat,Train_lab, test_size=0.33, random_state=42)
```

```python
print('{} items total, {} true'.format(X_train.shape, y_train.shape))
print('{} items total, {} true'.format(X_test.shape, y_test.shape))
```

```python
def auc(y_true, y_pred):
    auc = tf.metrics.auc(y_true, y_pred)[1]
    K.get_session().run(tf.local_variables_initializer())
    return auc
```

```python
classifier = Sequential()
#First Hidden Layer
classifier.add(Dense(1000, activation='tanh', kernel_initializer='normal', input_dim=37))
classifier.add(Dropout(0.50, noise_shape=None, seed=None))
#Second  Hidden Layer
classifier.add(Dense(526, activation='tanh', kernel_initializer='normal'))
classifier.add(Dropout(0.13, noise_shape=None, seed=None))
#Third Hidden Layer
classifier.add(Dense(128, activation='relu', kernel_initializer='normal'))
classifier.add(Dropout(0.09, noise_shape=None, seed=None))
#Fourth Hideen Layer
classifier.add(Dense(32, activation='relu', kernel_initializer='normal'))
classifier.add(Dropout(0.09, noise_shape=None, seed=None))
#Output Layer
classifier.add(Dense(1, activation='sigmoid', kernel_initializer='normal'))
#Cost Function metrics as AUC under ROC curve
classifier.compile(loss='binary_crossentropy', optimizer='adam', metrics=[auc])
```

```python
classifier.fit(X_train,y_train, epochs=1000, batch_size=128, verbose=0,  validation_data=(X_test,y_test))
```
